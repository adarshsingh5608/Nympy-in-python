```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 15000, 18000, 22000, 25000],  
    [2, 12000, 14000, 16000, 19000],  
    [3, 20000, 23000, 26000, 30000],  
    [4, 18000, 21000, 24000, 27000],  
    [5, 16000, 18000, 20000, 23000],
    [6, 18000, 19000, 26000, 23000]
])
```

```python
    print("sales of first three resturent", sales_data[1:3])
```

```python
year_sum = np.sum(sales_data[:,1:],axis=0)

print("Total sale of resturent per year are:", year_sum)
```

```python

yeary_total = np.sum(sales_data[:, 1:], axis=1)
print(yeary_total)
```

```python
min_sale = np.min(sales_data[:,1:],axis=1)
print("minimum sale of each resturant is : ", min_sale)
```

```python
max_sale = np.max(sales_data[:,1:],axis=1)
print("maximum sale of resturant is : ",max_sale)
```

```python

cumsum = np.cumsum(sales_data[:, 1:], axis=1)
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Sales")
plt.xlabel("Years")      # Fixed: Xlable -> xlabel
plt.ylabel("Sales")      # Fixed: Ylable -> ylabel
plt.grid(True)          # Fixed: false -> False
plt.show()
```

```python
# sorting in array 
```

```python
unsorted = np.array([4,3,5,1,2,4,6,7,4])
arr= np.sort(unsorted)
print("\n")
print(arr)
a2 = np.array([[3,1],[1,2],[2,3]])
print(np.sort(a2,axis=0))
print("\n")
print(np.sort(a2,axis=1))
```

```python
#adding rows and column in 2d array
```

```python
orw = np.array([[1,2],[3,4]])
new_row = np.array([[5,6]])
newar = np.vstack((orw,new_row))
print(newar)
print("\n")
new_col = np.array([[7],[8]])
newcol = np.hstack((orw,new_col))
print(newcol)
```
